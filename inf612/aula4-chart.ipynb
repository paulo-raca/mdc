```r
p <- ggplot(faithful, aes(x = waiting, y=eruptions))
p <- p + geom_point()
p <- p + geom_line()
p
```

```r
a <- -10:10
f1 <- a + 5
f2 <- (a^2) -a -5
df <- data.frame(a, f1, f2)

p <- ggplot(df, aes(x = a))
p <- p + geom_point(aes(y=f1))
p <- p + geom_line(aes(y=f2))
p
```

```r
air <- data.frame(year=as.numeric(time(AirPassengers)), num=matrix(AirPassengers))
p <- ggplot(air, aes(x = year))
p <- p + geom_line(aes(y=num))
p
```
