```r
unique(iris$Species)
#iris
iris2d=prcomp(iris[,-5])$x[,1:2]

labels = list(setosa=1, versicolor=2, virginica=3)
labels = as.character(labels[iris$Species])

clusters <- hclust(dist(iris2d), method = 'ward.D')
plot(clusters, labels=labels, cex = 0.3, hang=-1)

rect.hclust(clusters, k = 3)

table(labels, cutree(clusters, 3))
```

```r
iris2d=prcomp(iris[,-5])$x[,1:2]
ii=iris2d[1:20,]
dd=dist(ii)
h.c=hclust(dd,"complete")
plot(h.c,main="Complete")
plot(h.c,main="Complete",hang=-1)
```
